```python
from nltk.corpus import stopwords
stopwords.words('spanish')
```

```python
tokens=word_tokenize(miTexto,"spanish")
clean_tokens = ""
sr = stopwords.words('spanish')
for token in tokens:
    if token.lower() not in sr:
        clean_tokens.join(token)
```

```python
import pandas as pd

file = ('/content/drive/MyDrive/Colab Notebooks/MORS 2324/trial_es.tsv')
```

```python
data = pd.read_csv(file, delimiter='\t', usecols =["text","HS"])
```

```python
data.head
```

```python
data=data[:10]
data
```

```python
vocab=[]
for index, row in data.iterrows():
  for word in nltk.word_tokenize(row['text']):
    vocab.append(word)
```

```python
sp_stemmer = SnowballStemmer('spanish')
clean_vocab=[]
for word in vocab:
  if(word.isalnum()): #elimino simbolos
    if word.lower() not in stopwords.words('spanish'):
      clean_vocab.append(sp_stemmer.stem(word.lower()))
print(clean_vocab)
```

```python
from nltk.probability import FreqDist
fdist=FreqDist(clean_vocab)
fdist.keys()
fdist.plot(15, cumulative=False)
```
